## Function that extracts the given percentage of values from the dataset

The selection functions takes the dataframe as an argument and randomly selects the the number of rows based on the percentage and appends it to a new dataframe called "data".

```python
data = pd.DataFrame()
def selection(df):
    global data
    for value in df["db_id"].unique():
        val = df.loc[df["db_id"] == value]
        rows = val.sample(n = int(round((len(val) * 0.103) , 0)) )
        data = data.append(rows)
```

```python
selection(df_dev)
```

```python
selection(df_others)
```

```python
selection(df_spider)
```

### Adding of 1 row to make it to 1000

Since the selection fucntions selects the data only from those db_id values that are of length greater than 4, hence the db_id: "real_estate_properties" was dropped by the function which is then added manually to make the whole 1000 rows dataset.

```python
val = df_dev.loc[df_dev["db_id"] == 'real_estate_properties']
rows = val.sample()
data = data.append(rows)
```

### Analyzing Frequency of length of question in the dataset 

Appending the length of each question in the dataset to plot the graph of their frequency.

```python
len(df_dev["question_toks"])
```

```python
def string_length(str):
    x = len(str)
    return x

data['question_toks_length'] = data['question_toks'].apply(lambda x: string_length(x))
```

```python
data
```

```python
x = data["question_toks_length"]
plt.figure(figsize=(8,8))
plt.hist(x, bins=10)
plt.xlabel('Length of Questions')
plt.ylabel('Frequency of Questions')
plt.savefig( "Length-of-questions-freq.pdf" , dpi=1200)
plt.show()
```
